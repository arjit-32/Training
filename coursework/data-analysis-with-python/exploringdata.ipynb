```python
import pandas as pd;
```

```python
l = pd.DataFrame({'col1': [1,2,3,4,5], 'col2':['a','b','c','d','e'] })
r = pd.DataFrame({'col1': [3,4,5,6,7], 'col3':['f','g','h','i','j'] })
print(l)
print(r)
```

```python
# inner join 2 dataframes
pd.merge(left=l,right=r, on='col1', how='inner')
```

```python
pd.merge(left=l,right=r, on='col1', how='left')
```

```python
pd.merge(left=l,right=r,on='col1',how='right')
```

```python
# does a full merge
pd.merge(left=l,right=r, on='col1', how='outer')
```

```python
df1= pd.DataFrame({'account_id':[1,2,3,11382], 
                    'gender':['female','male','female','male'],
                    'age':[55,25,29,39]})
df1
```

```python
df2= pd.DataFrame({'account_id':[4,5,6,7], 
                    'gender':['female','male','female','male'],
                    'age':[19,28,14,15]})
df2
```

```python
# can concat data between 2 tables rowwise or columnwise
pd.concat([df1,df2], ignore_index=True)
```
