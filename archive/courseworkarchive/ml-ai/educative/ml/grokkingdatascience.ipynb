```python
# Computations and Agregations 

# Computations 

# Numpy's ufuncs (universal functions)
arr = np.arange(1,10)
print(np.add(arr, 2))
print(np.subtract(arr, 2))
print(np.multiply(arr, 2))
print(np.divide(arr, 2))
print(np.power(arr, 2))
print(np.mod(arr, 2))

# Exponential and Logs
print(np.exp(arr))
print(np.log(arr))
```

```python
# Reduce method 
arr = np.arange(1,10) 
print(np.add.reduce(arr)) # apply some operation to reduce an array to a single value

# Accumulate method
print(np.add.accumulate(arr)) # Keeps track of the intermediate results of the computation
```

```python
# Aggregations
x = np.random.random(100)

print(np.sum(x))
print(np.min(x))
print(np.max(x))
print(np.mean(x))
```

```python
# Comparisons
x = np.array([1,2,3,4,5])

print(x < 3)
print(2*x == x**2)
print(np.array_equal(x, x))
print(np.any(x > 3)) # True if any element in the array is True
print(np.all(x > 3)) # True if all elements in the array are True
```

## 2. Fundamental of Statistics 

### Basic Concepts 
- Mean: Sum the elements divided by the number of elements
- Median: The middle value of the sorted list
- Mode: The most frequent value in the list
- Variance: The average of the squared differences from the Mean
- Standard Deviation: The square root of the variance. (How much is the data spread out from the mean)
- Covariance: The measure of how two variables change together
- Correlation: The measure of how two variables change together divided by the standard deviation of each variable. If the correlation is positive, the variables change in the same direction. If the correlation is negative, the variables change in opposite directions.

Note - Correlation is not causation. i.e. Just because two variables are correlated, it does not mean that one causes the other.

### Basics of Probability
Probability is measure of likelihood of an event occuring

a. Independent Events: Outcome of one doesnt affect other
    P(X and Y) = P(X) * P(Y)

b. Dependent Events: Outcome of one affects other
    P(X and Y) = P(X) * P(Y after X has occured)

c. Mutually Exclusive Events: Events that cannot occur at the same time
    P(X or Y) = P(X) + P(Y)

d. Inclusive Events: Events that can occur at the same time
    P(X or Y) = P(X) + P(Y) - P(X and Y)

e. Conditional Probability: Probability of an event occuring given that another event has already occured
    P(X|Y) = P(X and Y) / P(Y)

### Bayesian Statistics

Bayes Theorem: P(A|B) = P(B|A) * P(A) / P(B)

Example: Want to find out a patient’s probability of having an allergy
- For people that really do have the allergy, the test says “Positive” 80% of the time, (Positive|Allergy)
- The probability of the test saying “Positive” to anyone is 10%, P(Positive)
If 1% of people have the allergy, P(Allergy), what is the probability that a person who tests positive actually has the allergy, P(Allergy|Positive)?

P(Allergy|Positive) = P(Positive|Allergy) * P(Allergy) / P(Positive)

### Probability Distributions

- *Uniform Distribution:* Values have same probability of occuring within a range, all values outside the range have zero probability of occuring. [Use for Categorical variable*]
- Bernoulli Distribution: When we have single trial with two possible outcomes, success or failure
- Binomial Distribution: When we have multiple trials with two possible outcomes, success or failure
- *Normal(Gaussian DIstribution):* Symmetric distribution where most of the values are close to the mean and few values are far away from the mean
- Poisson Distribution: The number of times an event occurs in an interval of time or space
- Exponential Distribution: The time between events in a Poisson process


## 3. Machine Learning 101

Types of ML algorithms - 
- Supervised Learning: The algorithm learns from labeled data. It can be either classification or regression problem. Ex- Linear Regression, SVM, Logistic Regression, Decision Tree, Random Forest, KNN, Neural Networks.

- Unsupervised Learning: The algorithm learns from unlabeled data. It can be either clustering or association problem. Ex- K-Means, Hierarchical Clustering, Apriori, EM, PCA.

- Semi-supervised Learning: The algorithm learns from a combination of labeled and unlabeled data. Ex- Google Photos, Google News.

- Reinforcement Learning: The algorithm learns by interacting with the environment. Ex- Markov Decision Process, Q-Learning.

### ML Algorithms 

1. Linear Regression - Finding a line/curve with best fit for the data points. Can be simple or multiple.

2. Logistic Regression - Used for classification problems. The output is a probability between 0 and 1. If the probability is greater than 0.5, the data point is classified as 1, else 0.

3. Decision Tree - A tree like structure where each node represents a feature, each link represents a decision rule and each leaf represents an outcome.

4. Naive Bayes - Based on Bayes Theorem. It assumes that the presence of a particular feature in a class is unrelated to the presence of any other feature. Ex- Spam Detection, Sentiment Analysis, Recommendation System.

5. SVM - Used for classification problems. It finds a hyperplane that best separates the data points into different classes.

6. KNN - Used for classification problems. It classifies a data point based on the majority of its neighbors. It is a lazy learner.

7. K-Means - Used for clustering problems. It divides the data points into k clusters such that the data points in the same cluster are similar and data points in different clusters are dissimilar.

8. Random Forest - Used for classification and regression problems. It creates multiple decision trees and combines them together to get a more accurate and stable prediction.

9. Dimensionality Reduction - It reduces the number of features in the data by combining the features together. Ex- PCA, LDA, Kernel PCA.

10. Gradient Boosting - It is an ensemble technique where the predictions of multiple weak models are combined together to get a more accurate and stable prediction. Ex- XGBoost, LightGBM, CatBoost.

11. Neural Networks - It is a network of interconnected neurons that are used to solve complex problems. Ex- CNN, RNN, LSTM, GAN.

### Evaluating a Model

1. Confusion Matrix - It is a table that is used to evaluate the performance of a classification model. It shows the number of correct and incorrect predictions made by the classification model compared to the actual outcomes. 

What percentage of the positive predictions were correct? - Precision

What percentage of the actual positive cases did the model catch? - Recall

2. ROC Curve - It is a graph that shows the performance of a classification model at all classification thresholds. It plots the True Positive Rate (TPR) against the False Positive Rate (FPR) at various threshold values.
