```python
# Creating a new column from Series
car_sales['Seats'] = pd.Series([5, 5, 5, 5, 5]) # Create a new column
car_sales
```

```python
car_sales['Seats'].fillna(5, inplace=True) # Fill missing values with 5
car_sales
```

```python
# New column from python list has to be same size as dataframe
fuel_economy = [7.5, 9.2, 5.0, 9.6, 8.7, 4.7, 7.6, 8.7, 3.0, 4.5]
car_sales['Fuel per 100KM'] = fuel_economy # Create a new column
car_sales
```

```python
car_sales['Total fuel used (L)'] = car_sales['Odometer (KM)'] / 100 * car_sales['Fuel per 100KM'] # Create a new column
car_sales
```

```python
car_sales['Passed road safety'] = True # Create a new column
car_sales
```

```python
# Dropping a column
car_sales.drop('Total fuel used (L)', axis=1, inplace=True) # axis=1 refers to columns, axis=0 refers to rows
car_sales
```

```python
car_sales["Odometer (Miles)"] = car_sales["Odometer (KM)"].apply(lambda x: x / 1.6) # Create a new column
car_sales
```
